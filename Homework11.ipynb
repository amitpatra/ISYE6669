```python
import cvxpy as cp
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = cp.Variable(2)
c = np.array([12, 1])

obj = cp.Minimize(c.T @ x)
constr = [
    20 * x[0] + x[1] >= 30,
    x[1] >= 20,
    x[0] >= 0
]

prob = cp.Problem(obj, constr)
prob.solve()

print("The optimal objective function value is: {}".format(prob.value))
print("The optimal solution is:")
print("y = {}".format(x[0].value))
print("r = {}".format(x[1].value))
```

```python
img = np.loadtxt("clownImage.txt")
plt.gray()

U, S, V = np.linalg.svd(img)
V = V.T
k = np.array([5, 15, 25])

for i in k:
    Xk = U[:, 0: i].dot(np.diag(S[0: i], 0))
    Xk = Xk.dot(V[:, 0: i].T)
    plt.imshow(Xk)
    print("k = {0}".format(i))
    plt.show()
```
